```python
import os
import json
from collections import defaultdict

# --- CONFIGURATION ---
RESULTS_DIR = "results/Exp1_Baseline/20251207_22_56_39"

# Define the full, ordered list of websites to ensure they always appear
WEBSITE_ORDER = [
    "Allrecipes", "Amazon", "Apple", "ArXiv", "BBC News", "Booking", 
    "Cambridge Dictionary", "Coursera", "ESPN", "GitHub", "Google Flights", 
    "Google Map", "Google Search", "Huggingface", "Wolfram Alpha"
]

def calculate_accuracy():
    if not os.path.exists(RESULTS_DIR):
        print(f"Error: Results directory not found at '{RESULTS_DIR}'")
        return

    tasks_by_website = defaultdict(lambda: {'total': 0, 'success': 0})

    print(f"\nAnalyzing results in: {RESULTS_DIR}")

    # 1. Gather data from summary files
    for root, dirs, files in os.walk(RESULTS_DIR):
        if "summary.json" in files:
            summary_path = os.path.join(root, "summary.json")
            try:
                with open(summary_path, 'r', encoding='utf-8') as f:
                    summary_data = json.load(f)
                
                task_id = summary_data.get("task_id", "Unknown-Task")
                status = summary_data.get("status", "Failed")
                website_name = task_id.split('--')[0]
                
                # Only count websites from our official list
                if website_name in WEBSITE_ORDER:
                    tasks_by_website[website_name]['total'] += 1
                    if status == "Success":
                        tasks_by_website[website_name]['success'] += 1
            except Exception as e:
                print(f"Warning: Could not parse {summary_path}. Error: {e}")

    # --- 2. Prepare Data for the Table ---
    table_data = []
    overall_total = 0
    overall_success = 0

    # Iterate through the FIXED list to ensure order and inclusion of all sites
    for name in WEBSITE_ORDER:
        stats = tasks_by_website[name] # This will default to 0/0 if not found
        total = stats['total']
        success = stats['success']
        
        if total > 0:
            rate = (success / total) * 100
            rate_str = f"{rate:.1f}%"
        else:
            rate_str = "0.0%" # Show 0.0% instead of N/A for consistency
            
        table_data.append([name, str(success), str(total), rate_str])
        
        overall_total += total
        overall_success += success

    # --- 3. Calculate Overall Statistics ---
    if overall_total > 0:
        overall_rate = (overall_success / overall_total) * 100
        overall_rate_str = f"{overall_rate:.1f}%"
    else:
        overall_rate_str = "0.0%"
        
    table_data.append(['---' * 5, '---', '---', '---'])
    table_data.append(['Overall', str(overall_success), str(overall_total), overall_rate_str])

    # --- 4. Print the Elegant Table ---
    headers = ["Website", "Success", "Total", "Accuracy"]
    
    col_widths = [len(h) for h in headers]
    for row in table_data:
        for i, cell in enumerate(row):
            if len(cell) > col_widths[i]:
                col_widths[i] = len(cell)

    header_str = " | ".join(headers[i].ljust(col_widths[i]) for i, h in enumerate(headers))
    print("\n--- Agent Performance Report ---")
    print(header_str)
    print("-" * len(header_str))

    for row in table_data:
        row_str = " | ".join(cell.ljust(col_widths[i]) for i, cell in enumerate(row))
        print(row_str)

    print("\n--- End of Report ---")

if __name__ == "__main__":
    calculate_accuracy()
```
